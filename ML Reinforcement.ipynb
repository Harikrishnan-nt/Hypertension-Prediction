## Statistical Testing

- #### Independent T-test (1)
- #### One-Way ANOVA (2)
- #### Chi-Square (3)

### BMI Between Has_Hypertension (1)

```python
from scipy.stats import ttest_ind, levene
import numpy as np

yes_group = hp[hp['Has_Hypertension'] == 'Yes']['BMI'].dropna()
no_group = hp[hp['Has_Hypertension'] == 'No']['BMI'].dropna()

stat_lev, p_lev = levene(yes_group, no_group)

equal_var = p_lev > 0.05
t_stat, p_value = ttest_ind(yes_group, no_group, equal_var=equal_var)

print(f'\nT-statistic: {t_stat:.4f}')
print(f'P-value: {p_value:.4f}')
print(f'Test Type: {'Equal variance' if equal_var else 'welch'}')

if p_value < 0.05:
    print(f'People with Hypertension have {'higher' if yes_group.mean() else 'lower'} BMI')
else:
    print(f'BMI is similiar between hypertension groups')
```

### Age Between Has_Hypertension (1)

```python
from scipy.stats import ttest_ind , levene
import numpy as np

yes_group = hp[hp['Has_Hypertension'] == 'Yes']['Age'].dropna()
no_group = hp[hp['Has_Hypertension'] == 'No']['Age'].dropna()

stat_lev, p_lev = levene(yes_group, no_group)

equal_var = p_lev > 0.05
t_stat, p_value = ttest_ind(yes_group, no_group, equal_var=equal_var)

print(f'\nT-statistic: {t_stat:.4f}')
print(f'P-value: {p_value:.4f}')
print(f'Test Type: {'Equal variance' if equal_var else 'welch'}')

if p_value < 0.05:
    print(f'People with Hypertension are {'older' if yes_group.mean() > no_group.mean() else 'younger'}')
else:
    print(f'Age is similiar between hypertension groups')
```

### Age Across Exercise_Level (2)

```python
from scipy.stats import f_oneway, levene
import numpy as np
import pandas as pd

low = hp[hp['Exercise_Level'] == 'Low']['Age'].dropna()
moderate = hp[hp['Exercise_Level'] == 'Moderate']['Age'].dropna()
high = hp[hp['Exercise_Level'] == 'High']['Age'].dropna()

stat_lev, p_lev = levene(low, moderate, high)
print(f"   p-value = {p_lev:.4f} {'(Equal variances)' if p_lev > 0.05 else '(Unequal variances)'}")

f_stat, p_value = f_oneway(low, moderate, high)

print(f"\nF-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print(f"  Age differs significantly across exercise levels")
    print(f"  Further post-hoc tests recommended to identify specific differences")
else:
    print(f"  Age is similar across all exercise levels")
```

### Salt_Intake Across BP_History (2)

```python
from scipy.stats import f_oneway, levene
import numpy as np
import pandas as pd

normal = hp[hp['BP_History'] == 'Normal']['Salt_Intake'].dropna()
prehypertension = hp[hp['BP_History'] == 'Prehypertension']['Salt_Intake'].dropna()
hypertension = hp[hp['BP_History'] == 'Hypertension']['Salt_Intake'].dropna()

stat_lev, p_lev = levene(normal, prehypertension, hypertension)
print(f"   p-value = {p_lev:.4f} {'(Equal variances)' if p_lev > 0.05 else '(Unequal variances)'}")

f_stat, p_value = f_oneway(normal, prehypertension, hypertension)

print(f"\nF-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print(f"  Salt intake differs significantly across BP history groups")
    print(f"  Further post-hoc tests recommended to identify specific differences")
else:
    print(f"  Salt intake is similar across all BP history groups")
```

### Has_Hypertension vs Smoking_Status (3)

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(hp['Has_Hypertension'], hp['Smoking_Status'])

chi2, p_value, dof, expected_freq = chi2_contingency(contingency_table)

print(f"Contingency Table: {contingency_table}")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencie: {expected_freq}")

if p_value < 0.05:
    print(f"  There is a significant relationship between hypertension and smoking status")
else:
    print(f"  There is no significant relationship between hypertension and smoking status")
```

### Model Building

 #### Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

le = LabelEncoder()
for col in ['BP_History', 'Medication', 'Family_History', 'Exercise_Level', 'Smoking_Status']:
    hp[col] = le.fit_transform(hp[col])

x = hp.drop('Has_Hypertension', axis=1)    
y = hp['Has_Hypertension'].map({'Yes':1, 'No':0})

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

model = LogisticRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf', C=1, gamma='scale')
svm_model.fit(x_train, y_train)

y_pred = svm_model.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

### K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)

y_pred = dt.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

### Evaluating Models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

models = {
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(kernel='rbf', C=1, gamma='scale'),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}

results = []

for name, model in models.items():
    if name in ["Logistic Regression", "SVM", "KNN"]:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
    else:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    results.append([name, acc, prec, rec, f1, cm])

results = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'Confusion Matrix'])    
print(results)
```

### Hypertuning the most accurate model

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_deepth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3]
}

grid = GridSearchCV(
    estimator=XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid.fit(x_train, y_train)
print('Best Parameters:', grid.best_params_)
best_xgb = grid.best_estimator_
```

```python
y_pred_best = best_xgb.predict(x_test)

print('Accuracy After Tuning:', accuracy_score(y_test, y_pred_best))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_best))
print('Classification Report:\n', classification_report(y_test, y_pred_best))
```
